## Anagram

```python
def validAnagram(arr1, arr2):
    # If both lengths are not equal they are false
    if len(arr1) != len(arr2):
        return False
    
    # Define two object
    FreqArr1 = {}
    FreqArr2 = {}
    
    # Dict object of arr1
    for i in arr1:
        if i not in FreqArr1:
            FreqArr1[i] = 1
        else:
            FreqArr1[i] += 1
    
    # Dict object of arr2
    for i in arr2:
        if i not in FreqArr2:
            FreqArr2[i] = 1
        else:
            FreqArr2[i] += 1
            
    
    # Compare both object's items key value pair should be equal
    for key in FreqArr1.keys():
        if (key not in FreqArr2) or (FreqArr1[key] != FreqArr2[key]):
            return False
    
    return True
```

```python
validAnagram(text1, text2)
```

## countUniqueValues

```python
def countUnique(arr):
    if len(arr) < 1:
        return 0
    
    # First define two pointers of adjacent array index
    # Set two pointers i starts at 0, j starts at 1
    i = 0
    j = 1

    # While the pointer is less than array length
    while j < len(arr):
        # Compare i and j
        # If same then j + 1
        if arr[i] == arr[j]: 
            j += 1

        else:
        # If not same then i + 1 and arr[i] = arr[j]
            i += 1
            arr[i] = arr[j]

    # Then return the value of i + 1
    return i + 1
```

# Section 8 Recursion Problem

## power

```python
## Example
# pow(2, 0) = 1
# pow(2, 2) = 4
# pow(2, 4) = 16
```

```python
def poww(x, p):
    # Create base case
    if p == 0:
        return 1
    
    # Create recursion
    return x * pow(x, p-1)
```

```python
print(poww(2, 0))
print(poww(2, 2))
print(poww(2, 4))
```

## recursiveRange
Create function that adds up number from `0` to number passed into the function

```python
# SAMPLE INPUT/OUTPUT
# recursiveRange(6) // 21
# recursiveRange(10) // 55 
```

```python
def recursiveRange(n):
    # Base case
    if n == 1:
        return 1
    # Recursion
    return n + recursiveRange(n-1)
```

```python
print(recursiveRange(6))
print(recursiveRange(10))
```

## Binary Search

```python
# # Create function that accepts sorted array, return index at which value exist
# Example:
# binarySearch([1, 2, 3, 4, 5], 2) // 1
# binarySearch([1, 2, 3, 4, 5], 3) // 2
# binarySearch([1, 2, 3, 4, 5], 6) // -1
```

```python
from math import floor                 
def binarySearch(arr, val):
    # First define start middle and end index
    sidx = 0
    eidx = len(arr)-1
    midx = floor(sidx + eidx/2)

    # While arr[cidx] != val and sidx < eidx
    while((arr[midx] != val) and (sidx < eidx)):
        #print(f"{sidx} {midx} {eidx}")
        
        if arr[midx] > val:
            eidx = midx - 1

        elif arr[midx] < val: 
            sidx = midx+1

        midx = floor((eidx+sidx)/2)
        
    # return index or -1
    if arr[midx] == val:
        return midx

    return -1
```

# Section 11: Bubble Sort

```python
# Bubble sort without optimization
# Example
# bubbleSort([37, 45, 29, 8]) // 8, 29, 37, 45
```

```python
# Createa swap function
def swap(val1, val2):
    temp = val1
    val1 = val2
    val2 = temp
    
    return val1, val2
```

```python
# Test swap
a = 5
b = 20
a, b = swap(a, b)

print(f"val of a:{a}, b:{b}")
```

```python
# Optimized add, the noSwaps boolean flag to indicate if swap isn't happening
# If there's no swap then array already sorted and no need to do the rest of process

def bubbleSort(arr):    
    # Iterate through all arr elements, from last index to start index
    for i in range(len(arr))[::-1]:
        # No swap flag
        noSwap = True
        
        # print('arr:', arr)
        # Iterate from j=0 to i-1
        for j in range(i):
            
            # print(f"{arr[j]} and {arr[j+1]}")
            # Compare arr[j] and arr[j+1]
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = swap(arr[j], arr[j+1])
                noSwap = False
        
        # If no swap occurs, then get out of loop
        if noSwap:
            break
    
    return arr
```

```python
# first start at second index i iterate through the whole arr
# Note that the udemy course in this topic, is not really well explained, look for youtube tutorial instead
def insertionSort(arr):
    for i in range(1, len(arr)):

        # Select the value to be compared 
        currentVal = arr[i]

        #print(f'arr : {arr}, curr val {currentVal}')
        # While loop comparing arr[i] and arr[i-1], swap if arr[i-1] > arr[i]
        while arr[i-1] > arr[i] and i != 0:
            #print('swap')
            arr[i], arr[i-1] = arr[i-1], arr[i]
            i -= 1

        #print(f'after one pass {arr}')
        
    return arr
```

## `merge`

```python
def merge(arr1, arr2):
    # Create empty arr to store result
    res = []

    # initialize the index of arr1 and arr2
    i = 0
    j = 0

    while (i < len(arr1)) and (j < len(arr2)):
        # When arr[i] <= arr2[j]
        if arr1[i] <= arr2[j]:
            res.append(arr1[i])
            i += 1
        
        # When arr[i] > arr2[j]
        else:
            res.append(arr2[j])
            j += 1
    
        # print(f"Current res {res}")
    
    # print(f'Current i : {i}')
    # print(f'Current j : {j}')

    # When either one is exhausted
    while i < len(arr1): # This means arr2 is exhausted
        # print('The arr2 is exhausted')
        res.append(arr1[i])
        i += 1

    while j < len(arr2): # This means arr1 is exhausted
        # print('The arr1 is exhaussted')
        res.append(arr2[j])
        j += 1

    return res
```

```python
def mergeSort(arr):
    half = len(arr)//2
    if len(arr) <= 1:
        return arr
    left = mergeSort(arr[:half])
    right = mergeSort(arr[half:]) 
    return merge(left, right)
```

```python
# Create swap that receives array and two indices to be swapped
def swap(arr, idx1, idx2):
    temp      = arr[idx1]
    arr[idx1] = arr[idx2]
    arr[idx2] = temp
```

## `pivot`

```python
def pivot(arr, start=0, end=len(arr)):
    # Define start idx, st = 0 
    stix = 0

    # Define the current pivot index
    pvix = 0

    # Set the first element to be pivot
    pivot = arr[0]
    #print(f"Initial: {arr}")
    
    # Iterate from pvix + 1 to length of array
    for i in range(stix+1, len(arr)):
        if pivot >= arr[i]:
            pvix += 1
            swap(arr, i, pvix)
            # print(f"Swapping!: {arr}")
        
    # Lastly swap pivot to pvi index
    swap(arr, stix, pvix)
    # print(f'Last step!: {arr}')  
    
    return pvix
```

## Helper function `getDigit`

```python
# Example
# getDigit(12345, 0) // 5
# getDigit(12345, 1) // 4
# getDigit(12345, 2) // 3
# getDigit(12345, 3) // 2
# getDigit(12345, 4) // 1
# getDigit(12345, 5) // 0
```

```python
# Return the digit in num at the given place value
def getDigit(num, place):
    return (num//(10**place))%10    
```

## `mostDigits(nums)`

```python
# Given an array of numbers returns number of digits in largest numbers in the list
# mostDigits(nums)
# mostDigits([1234, 56, 7]); //4
# mostDigits([1, 1, 11111, 1]); //5
# mostDigits([12, 34, 56, 78]); //2
```

```python
def mostDigits(nums):
    most = 0    
    for n in nums:
        most = max(digitCount(n), most)
    
    return most 
```

```python
mostDigits([1234, 334564, 56, 78])
```

## `radixSort`

```python
def radixSort(arr):
    # Figure out the largest number
    n_iter = mostDigits(arr)

    for k in range(n_iter):
        # print(f'Right now k is {k}')

        # Create bucket for each digit
        bucket = []

        # Iterate over arr, getDigit equal to k
        for i in range(9):
            bucket += [val for val in arr if getDigit(val, k) == i]

        # Assign newly sorted newarr as arr
        arr = bucket

    return arr
```
